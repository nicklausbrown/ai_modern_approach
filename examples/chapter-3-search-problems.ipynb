```python
import random
import networkx as nx
```

```python
nodes = list(range(20))
edges = [random.sample([y for y in nodes if y != x], k=4) for x in nodes] # no self-reference in graph
graph = dict(zip(nodes, edges))
```

```python
G = nx.Graph()
for node, node_array in graph.items():
    for connected_node in node_array:
        G.add_edge(node, connected_node)
```

```python
nx.draw_networkx(G)
```
