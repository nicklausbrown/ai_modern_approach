```python
import random
import networkx as nx
```

```python
# Allow cycles in the graph, in fact - do nothing about their probability or connectedness 
```

```python
NODE_COUNT = 20
BRANCH_FACTOR = 2

nodes = list(range(NODE_COUNT))
edges = [random.sample([y for y in nodes if y != x], k=BRANCH_FACTOR) for x in nodes] # no self-reference in graph
graph = dict(zip(nodes, edges))
```

```python
G = nx.Graph()
for node, node_array in graph.items():
    for connected_node in node_array:
        G.add_edge(node, connected_node)
```

```python
nx.draw_networkx(G)
```
